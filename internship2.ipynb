```python
grocery_list = ["apple","banana","milk","egg","yogurt","bread"]

new_list = ["juice","pencil box","cheese","eggs","milk","pens"]

for item in new_list:
  if item in grocery_list:
    print(f"{item} already in the list")
  else:
    grocery_list.append(item)
    print(f"{item} add to the list")

print("updated grocery list:")
print(grocery_list)


```

```python
def calculate_total_bill(item_prices):
  platform_fee = 5
  delivery_charge = 80
  subtotal = sum(item_prices)
  subtotal += platform_fee
  subtotal += delivery_charge
  return subtotal

item_prices = {100,200,300,400}
subtotal = calculate_total_bill(item_prices)
if subtotal > 1000:
  reward= 200
elif subtotal >500:
  reward= 50
else:
  reward = 0
print(f"your subtotal is {subtotal} and reward is {reward}")
```
